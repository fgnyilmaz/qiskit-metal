```python
# For convenience, let's begin by enabling
# automatic reloading of modules when they change.
%load_ext autoreload
%autoreload 2

import numpy as np
from collections import OrderedDict

import warnings
warnings.filterwarnings('ignore')

import qiskit_metal as metal
from qiskit_metal import designs, draw
from qiskit_metal import MetalGUI, Dict
import matplotlib.pyplot as plt

from qiskit_metal.toolbox_metal import math_and_overrides

from qiskit_metal.qlibrary.core import QComponent


from qiskit_metal.qlibrary.qubits.grAl_test_structure import FluxoniumPocket
```

```python
# Each time you create a new quantum circuit design,
# you start by instantiating a QDesign class.

# The design class `DesignPlanar` is best for 2D circuit designs.
design = designs.DesignPlanar()
design.chips.main.material = 'silicon'
design.variables['cpw_width'] = '15 um'
design.variables['cpw_gap'] = '8.733 um'
design._chips['main']['size']['size_x'] = '9mm'
design._chips['main']['size']['size_y'] = '9mm'

# If you disable the next line with "overwrite_enabled", then you will need to
# delete a component [<component>.delete()] before recreating it.
design.overwrite_enabled = True

gui = MetalGUI(design)
```

```python
from qiskit_metal.qlibrary.qubits.grAl_test_structure import FluxoniumPocket
# TS1 details
test_struc = 0 #initial number for grAl test structure
iw = 0.01 #initial width of grAl inductor
p_x = 0.2

while iw < 0.1:
    options = dict(chip='main', 
                   pos_x = '-4.2mm' +str(p_x), pos_y = '4.2mm',
                  )
               
ts = FluxoniumPocket(design,'TS' +str(test_struc), options = dict(
        **options))

test_struc += 1
iw += 0.05

gui.rebuild()
```

```python
from qiskit_metal.qlibrary.qubits.grAl_test_structure import FluxoniumPocket
# TS1 details

options = dict(chip='main', 
               pos_x = '-2.8mm', pos_y = '2mm',
                )
               
ts1 = FluxoniumPocket(design,'TS1', options = dict(
        **options))


options = dict(chip='main', 
               pos_x = '-2.0mm', pos_y = '2mm',
            inductor_width='120um',
                )
               
ts2 = FluxoniumPocket(design,'TS2', options = dict(
        **options))

options = dict(chip='main', 
               pos_x = '-1.2mm', pos_y = '2mm',
            inductor_width='140um',
                )
               
ts3 = FluxoniumPocket(design,'TS3', options = dict(
        **options))

gui.rebuild()
```
